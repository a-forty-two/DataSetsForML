```
fl = open('ben.txt','rt')
data = fl.read()
import spacy
nlp = spacy.load('en_core_web_sm')
text = nlp(data) # TOKENIZATION!!! 1 step pipeline?
```

The magic behind this 1-step pipeline is an entire series of functions!


```
for i, entity in enumerate(text.ents):
  print(entity.text, entity.label_)
  if i>10:
    break
```

```
# CREDIT CARD NUMBERS, SSN, Aadhaar-> sensitive, PII, SPDI data, protected data 
# CANNOT BE SHARED
# Data cleaning process-> censor such data 
# let's take an example to censor all the names

def censor(content):
  if content.ent_type_ == "GPE": # FROM SPACY, REGEX -> to remove any custom pattern on content.text
    return " *** "
  else:
    return content.string

# apply this logic for entire data set [entire file]
def cleanData(inputData):
  doc = nlp(inputData)
  for ent in doc.ents:
    ent.merge()
  tokens = map(censor, doc)
  return "".join(tokens)

print(cleanData(data))
# https://spacy.io/api/annotation#named-entities 
```

```
# Extract detailed information about topics 
import textacy
statements = textacy.extract.semistructured_statements(text, "Benjamin")
# SUBJECT, VERB, FACT  <- statement
for statement in statements:
  subject, verb, fact = statement
  print("Subject: " + str(subject))
  print("verb: " + str(verb))
  print("Fact: " + str(fact))
```

```
# MULTIPLE Or Complex nouns are handled via NOUN CHUNKS
chunks = textacy.extract.noun_chunks(text, min_freq=1) # occurance at least thrice
# noun chunks are objects, and need to be converted to strings to be processed further
chunks = map(str, chunks) # [str(chunk) for chunk in chunks]
# convert all upper caps to lower case
#chunks = map(str.lower, chunks) # Ignore if you want to preserve the first letter
# but better practice is to convert everything to lower
# you could make first letter capital while printing any time when required
for chunk in set(chunks):
  if len(chunk.split(" ") )> 4: # controlling number of words to print
    print(chunk)

```

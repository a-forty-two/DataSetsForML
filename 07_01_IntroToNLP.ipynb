```
from collections import Counter
data = open('cogni_data.txt','rt')
content = data.read()
tokens = nlp(content)
print(tokens[0].vector)
```

```
# remove the stop words
all_words_before_stopremoval = [token.text for token in tokens]
print("length before removing stopwords: {}".format( len(all_words_before_stopremoval)))
words = [token.text for token in tokens if not token.is_stop and not token.is_punct]
print("length after removing stopwords: {}".format( len(words)))
print(words[:20])
```

```
# print some of the most common words
f1 = Counter(all_words_before_stopremoval)
f2 = Counter(words)
print(f1.most_common(5))
print(f2.most_common(5))
```

```
# unique words are the ones with frequency 1!!
unique1 = [word for (word,freq) in f1.items() if freq==1]
print(unique1) # without stop word removal
unique2 = [word for (word,freq) in f2.items() if freq==1]
print(unique2) # after stop word removal
```

```
# Can i identify segments/sequences of these tokens that can help me identify nouns and verbs and 
# other complex sequences? Words that contain ing are obvious verbs, ran is also a verb! 
# PART of SPEECH tagging -> breaks your sentences into logical grammatical groups
# tokens contain our original text
#tokens[:5]
#for token in tokens:
#  print(token.text, token.tag_, spacy.explain(token.tag_))
```

```
# grammatical filtering 
nouns = []
adj = []
for token in tokens:
  if token.pos_ == 'NOUN':
    nouns.append(token)
  if token.pos_ == 'ADJ':
    adj.append(token)

print(nouns[:5])
print(adj[:5])
```

```
from spacy import displacy 
# Directed Acyclic Graphs 
# TensorFlow and neural networks are not just for Deep Learning or ML, they are also for 
# other subjects like NLP 
# COMPUTATION and GRAPH calculation: chemistry, biology
displacy.render(tokens, style='dep')

```

```
for i,token in enumerate(tokens):
  print(token.text.lower())
  if i>5:
    break
```

```
corrected_words = []
for token in tokens:
  localword= token.text.lower()
  corrected_words.append(localword)
sentence = ' '.join(corrected_words)
new_data = nlp(sentence)
```

```
for i,token in enumerate(new_data):
  if i>10:
    break
  print(token.text, token.tag_, token.pos_,spacy.explain(token.tag_))
```
